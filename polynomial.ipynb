```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/apratim777/apratim777/master/Position_Salaries.csv")
df
```

```python
plt.xlabel('level')
plt.ylabel('salary')
plt.scatter(df.Level, df.Salary, color='red', marker='+')
```

```python
x = df.drop(['Position','Salary'], axis = 1)
x
```

```python
y = df.Salary
y
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
plt.xlabel('level')
plt.ylabel('salary')
plt.scatter(df.Level, df.Salary, color='red', marker='+')
plt.plot(x, reg.predict(x))
```

```python
from sklearn.preprocessing import PolynomialFeatures
ploy = PolynomialFeatures(degree = 10)
```

```python
x_poly  = ploy.fit_transform(x)
x_poly
```

```python
lin2 = linear_model.LinearRegression()
```

```python
lin2.fit(x_poly, y)
```

```python
lin2.predict(x_poly)
```

```python
lin2.score(x_poly,y)
```
